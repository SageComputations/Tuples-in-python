# 2. Tuples are enclosed in ( )

```python
int_tuple = (1,2,3)

print(int_tuple)
```

```python
str_tuple = ("Hello","Python")

print(str_tuple)
```

```python
print(int_tuple, str_tuple) #Print function can take in two tuples and will print them in sets of round brackets
```

```python
combined_tuple = int_tuple + str_tuple

combined_tuple
```

```python
print(str_tuple*3)
```

```python
mixed_tuple = (3,"Hello",4.4)

print(mixed_tuple)
```

# 3. Even when you leave the brackets, python correctly realises that it should be a tuple

```python
my_tuple = 3,"world",4.9

print(my_tuple)
```

# 4. Fields are used to acces elements in the tuple

```python
a,b,c = my_tuple
```

```python
print(a)
print(b)
print(c)
```

```python
nested_tuple = (1,2,3,(4,5,6)) #Tuples within tuples

print(nested_tuple)
```

```python
mixed_tuple = ("mouse",[8,4,6],(1,2,3))

print(mixed_tuple)
```

```python
my_tuple = ('p','e','r','m','i','t')

my_tuple
```

# 6. Tuples also support slicing operations

```python
my_tuple = ('G','i','r','a','f','f','e')

print(my_tuple[1:])
```

```python
print(my_tuple[2:4])
```

```python
print(my_tuple[::2])
```

```python
print(my_tuple[-1])
```

```python
my_tuple = ('H','e','l','l','o')
my_tuple
```

```python
print(my_tuple.index('H'))
```

```python
print('e' in my_tuple)
```
