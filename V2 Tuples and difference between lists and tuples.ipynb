# Tuples are immutable - tuples once created cannot be changed or updated in any way

```python
immutable_tuple_1 = (0,1,2,3)

immutable_tuple_1
```

```python
immutable_tuple_1[0] = 4

print(immutable_tuple_1)
```

```python
immutable_tuple_2 = (4,2,3,(6,5))

immutable_tuple_2
```

```python
immutable_tuple_2[0]
```

```python
immutable_tuple_2[0] = 9
```

```python
immutable_tuple_3 = (4,2,3,[6,5])

immutable_tuple_3
```

```python
immutable_tuple_3[3]
```

```python
immutable_tuple_3[3][1] = 100

print(immutable_tuple_3)
```

```python
my_tuple = ('c','h','r','i','s','t','m','a','s')

my_tuple
```

```python
del my_tuple[3]
```

```python
del my_tuple #We cannot delete the item of the tuple but can delete the tuple as a whole
```

```python
print(my_tuple)
```

```python
tuple_a = (1,2,3,4,5)
tuple_b = ('a','b','c','d','e')
```

# Zip function - makes pairs of elements of one tuple with corresponding elements of another

```python
zipped = zip(tuple_a,tuple_b) #Collection of tuples (1,'a')(2,'b')(3,'c'),(4,'d')(5,'e')

print(zipped) #Result of a zip function is an object 
```

```python
result = tuple(zipped)

print(result)
```

# Zipped result can be converted to a complex data type tuple,list,dict,set

```python
tuple_x, tuple_y = zip(*result)
```

```python
print('tuple_x: ',tuple_x)
print('tuple_y: ',tuple_y)
```

```python
list_a = [6,7,8,9,10]
list_b = ['a','e','i','o','u']
```

```python
zipped_list = zip(list_a,list_b)
```

```python
result = list(zipped_list)

result
```

```python
a = ('John','Charles','Mike',)

b = ('Manager','Supervisor','Engineer')

x = zip(a,b)
```

```python
print(dict(x))
```

```python
numbers_list = [1,2,3]
str_list = ['one','two','three']
numbers_tuple = ('ONE','TWO','THREE','FOUR')
result = zip(numbers_list,numbers_tuple)
```

```python
print(result)
```

```python
result_set = set(result)

print(result_set)
```
